# Feature Selection 

## Dataset 


Dataset "ABCtestscore.dta" is based on simplified version of the dataset collected as part of the field experiment in Niger that is now published in American Economic Journal: Applied Economics.

The paper is: Aker, J.C., Ksoll, C., & Lybbert, T.J. (2012). Can mobile phone improve learning? Evidence from a field experiment in Niger. American Economic Journal: Applied Economics, 4(4), 94-120.
The paper and the full dataset can be found here: https://www.aeaweb.org/articles?id=10.1257/app.4.4.94

Basically, the authors conducted a field experiment with a sample of adults participating to an education programme in Niger. They randomly allocated participants to either a treatment group (called ‘abc’) where students learned how to use simple mobile phones, or to a control group, with no such a training.
The paper looks at whether the mobile phone intervention had a treatment effect on the adults’ test scores in a math test (‘math’), and in a writing test (‘write’) in the education programme.

## VIF

```python
Y=data.abc
print (Y.sum())
X=data.drop(['abc'],axis=1) 
log_reg=linear_model.LogisticRegression()
log_reg.fit(X,Y)
# print (log_reg.score(X,Y))
# print (Y.mean())
```

```python
def calculate_vif_(X, thresh=5.0):
    variables = list(range(X.shape[1]))
    dropped = True
    while dropped:
        dropped = False
        vif = [variance_inflation_factor(X.iloc[:, variables].values, ix)
               for ix in range(X.iloc[:, variables].shape[1])]

        maxloc = vif.index(max(vif))
        if max(vif) > thresh:
            print('dropping \'' + X.iloc[:, variables].columns[maxloc] +
                  '\' at index: ' + str(maxloc))
            del variables[maxloc]
            dropped = True

    print('Remaining variables:')
    print(X.columns[variables])
    return X.iloc[:, variables]
```

```python
aa=calculate_vif_(X, thresh=5)
```

```python
log_reg=linear_model.LogisticRegression()
log_reg.fit(aa,Y)
data
print (log_reg.score(aa,Y))
print (Y.mean())
```

## Feature Importance 

```python
x_train1,x_test1,y_train1,y_test1=train_test_split(aa,Y)
log_reg2=tree.DecisionTreeClassifier()
log_reg2.fit(x_train1,y_train1)

print (y_train1.mean(), y_test1.mean()) 
print (len(y_train1), len(y_test1))
class_predict=log_reg2.predict(x_test1)
print (metrics.accuracy_score(y_test1,class_predict))
print (metrics.recall_score(y_test1,class_predict))
print (metrics.precision_score(y_test1,class_predict))
```

## Univariate Selection 

```python
bestfeatures = SelectKBest(score_func=chi2, k=10)
fit = bestfeatures.fit(aa,Y)
dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(aa.columns)
#concat two dataframes for better visualization 
featureScores = pd.concat([dfcolumns,dfscores],axis=1)
featureScores.columns = ['Specs','Score']  #naming the dataframe columns
print(featureScores.nlargest(10,'Score')) 
```
