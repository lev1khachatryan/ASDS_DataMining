```python
import numpy as np  
import matplotlib.pyplot as plt  
import pandas as pd  
from apyori import apriori 
```

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
data = [['Milk', 'Onion', 'Nutmeg', 'Kidney Beans', 'Eggs', 'Yogurt'],
           ['Dill', 'Onion', 'Nutmeg', 'Kidney Beans', 'Eggs', 'Yogurt'],
           ['Milk', 'Apple', 'Kidney Beans', 'Eggs'],
           ['Milk', 'Unicorn', 'Corn', 'Kidney Beans', 'Yogurt'],
           ['Corn', 'Onion', 'Onion', 'Kidney Beans', 'Ice cream', 'Eggs']]
```

```python
te = TransactionEncoder()
te_ary = te.fit(data).transform(data)
df = pd.DataFrame(te_ary, columns=te.columns_)
```

```python
frequent_itemsets_test = apriori(df, use_colnames=True)
frequent_itemsets_test
```

```python
frequent_itemsets = apriori(df, min_support=0.6, use_colnames=True)

frequent_itemsets
```

### Print the rules derived from the frequent itemsets only if the level of Confidence is above the 70 percent threshold (min_threshold=0.7) 

```python
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.7)
rules
```

### Print the rules derived from the frequent itemsets only if the level of Lift is above 1.2

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1.2)
rules
```

### Print the rules that satisfy the following criteria:

at least 2 antecedents,
the confidence > 0.75
and lift score > 1.2 

```python
rules["antecedent_len"] = rules["antecedents"].apply(lambda x: len(x))
rules
```

```python
rules[ (rules['antecedent_len'] >= 2) &
       (rules['confidence'] > 0.75) &
       (rules['lift'] > 1.2) ]
```
